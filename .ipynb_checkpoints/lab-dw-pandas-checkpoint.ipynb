### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
import pandas as pd
import numpy as np
import statistics as stats
```

**import data**

```python
customer = pd.read_csv('./customer.csv')
customer
```

```python
#convert titles to lower case and replace spaces with_
```

```python
columns = []
for col in customer.columns: 
    if ' ' in col:
        col = col.replace(' ','_')
        columns.append(col.lower())
    else:
        columns.append(col.lower())
        
customer.columns = columns
```

```python
customer = customer.dropna(axis = 0, how = 'all')
customer
```

```python
#Change Customer Lifetime Value column to be a float. Adding % to column name. 

customer['customer_lifetime_value'] = customer['customer_lifetime_value'].str.rstrip('%').astype('float') / 100.0
customer.rename(columns = {'customer_lifetime_value':'customer_lifetime_value_%'}, inplace = True)

customer['customer_lifetime_value_%'] = round(customer['customer_lifetime_value_%'],2)
```

```python
customer
```

```python
#In column Income and in Monthly Premium Auto,looks like all the floats are .0, we change to integers.

customer['income'] = customer['income'].astype(int)
customer['monthly_premium_auto'] = customer['monthly_premium_auto'].astype(int)
```

```python
customer.dtypes
```

```python
#In column Number of Open Complaints, looks like an integer is showed like a date. 1/00/00 should be 1. We change to an integer.

customer['number_of_open_complaints'] = customer['number_of_open_complaints'].str.split('/',n=2, expand=True)[1]
```

```python
customer['education'].unique()
```

```python
#In column Total Claim Amount, floats are too long. We change to just the first two decimal places. 

customer['total_claim_amount'] = round(customer['total_claim_amount'], 2)
```

```python
#We only want F and M so let's change the rest to make it uniform.
customer['gender'].value_counts()
```

```python
#we make a loop where errors are converted to M or F

genderok = []
for i in customer['gender']:
    #if Male append M
    if i =='Male':
        genderok.append('M')
    #elif female append F
    elif i =='female':
        genderok.append('F')
    #elif Femal append F
    elif i =='Femal':
        genderok.append('F')
    #else append i
    else:
        genderok.append(i)

#The 'gender' column now has the data of genderok
customer['gender'] = genderok
```

```python
customer['gender'].value_counts()
#here we check the unique values to know how we are going to fill the nans, with categories we put F (the most common).
```

```python
customer['gender'] = customer['gender'].fillna('F')
#then we replace nans with F.
```

```python
#We just want the full names of the states so let's change WA, Cali and AZ to make it uniform.
customer['st'].value_counts()
```

```python
#We make a loop where mistakes become Washington, California or Arizona.

stok = []
for i in customer['st']:
    #if WA append M
    if i =='WA':
        stok.append('Washington')
    #elif AZ append Arizona
    elif i =='AZ':
        stok.append('Arizona')
    #elif Cali append California
    elif i =='Cali':
        stok.append('California')
    #else append i
    else:
        stok.append(i)

#The column 'st' now has the stock data.
customer['st'] = stok
```

```python
customer['st'].value_counts()
#Here we check the unique values to know how we are going to fill the nans, with categories we put Oregon (the most common).
```

```python
customer['st'] = customer['st'].fillna('California')
#then we replace nans with California.
```

```python
#calculate from each numerical column: mean, median, mode, standard deviation, and quartiles
#numerical columns: customer_lifetime_value_%	income	monthly_premium_auto	number_of_open_complaints, total_claim_amount

customer[['customer_lifetime_value_%','income', 'monthly_premium_auto', 'total_claim_amount']].mean()
```

```python
numcol = ['customer_lifetime_value_%','income', 'monthly_premium_auto', 'total_claim_amount']
numcol

```

```python
for i in numcol:
    print(i)
    print('mean:', customer[i].mean())
    print('median:', customer[i].median())
    print('mode:',stats.mode(customer[i]))
    print('standard deviation:',stats.pstdev(customer[i]))
    print('quartiles:',stats.quantiles(customer[i], n=4, method='exclusive'))
    print(' ')
    print(' ')
```

*Note: When analyzing data, we often want to focus on certain groups of values to gain insights. Percentiles are a useful tool to help us define these groups. A percentile is a measure that tells us what percentage of values in a dataset are below a certain value. For example, the 75th percentile represents the value below which 75% of the data falls. Similarly, the 25th percentile represents the value below which 25% of the data falls. When we talk about the top 25%, we are referring to the values that fall above the 75th percentile, which represent the top quarter of the data. On the other hand, when we talk about the bottom 25%, we are referring to the values that fall below the 25th percentile, which represent the bottom quarter of the data. By focusing on these groups, we can identify patterns and trends that may be useful for making decisions and taking action.*
